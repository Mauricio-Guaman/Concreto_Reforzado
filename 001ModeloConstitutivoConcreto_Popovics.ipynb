```python
"""
Created on Tue Nov 12 15:27:32 2024

@author: Mauricio
"""
# Librerias
import numpy as np
import matplotlib.pyplot as plt
#%% Datos
k=2
fco_p=21; #[MPa]
Ec=4700*(fco_p)**0.5
A=900
eco=0.002
R=5

#%%  Concreto No Confinado
fcc_p=fco_p
ecc=eco*(1+R*(fcc_p/fco_p-1))
Esec=fcc_p/ecc
r=Ec/(Ec-Esec)

ec=np.linspace(0,0.003,100)
fc_list=[]
for ec_a in ec:
    x=ec_a/ecc
    fc=fcc_p*x*r/(r-1+x**r)
    print(fc)
    fc_list.append(fc)
fig=plt.figure(constrained_layout=True,figsize=(19/2,9.5/2))
gs=fig.add_gridspec(1,1)
# Ground motion
ax=fig.add_subplot(gs[0,0])
ax.plot(ec,fc_list,label='Concreto No Confinado')
#ax.plot(ec1,fc_list1,label='Concreto No Confinado')
ax.set_xlim(0,max(ec)+0.0005)
ax.set_ylim(0,max(fc_list)+2)
ax.grid(which='major', color='#DDDDDD', linewidth=0.8)
ax.grid(which='minor', color='#EEEEEE', linestyle='-', linewidth=0.5)
ax.minorticks_on()
ax.set(xlabel='Deformacin Unitaria',ylabel='Esfuerzo fc [MPa]')
ax.set_title('Modelo Constitutivo Concreto: Popovics')
plt.legend()
```

```python
#%% Concreto Confinado
fcc_p1=fco_p*k
ecc1=eco*(1+R*(fcc_p1/fco_p-1))
Esec1=fcc_p1/ecc1
r1=Ec/(Ec-Esec1)

ec1=np.linspace(0,0.018,100)
fc_list1=[]
for ec_a in ec1:
    x1=ec_a/ecc1
    fc1=fcc_p1*x1*r1/(r1-1+x1**r1)
    print(fc1)
    fc_list1.append(fc1)

#%% PLOT
fig=plt.figure(constrained_layout=True,figsize=(19/2,9.5/2))
gs=fig.add_gridspec(1,1)
# Ground motion
ax=fig.add_subplot(gs[0,0])
ax.plot(ec,fc_list,label='Concreto No Confinado')
ax.plot(ec1,fc_list1,label='Concreto No Confinado')
ax.set_xlim(0,max(ec1)+0.0005)
ax.set_ylim(0,max(fc_list1)+2)
ax.grid(which='major', color='#DDDDDD', linewidth=0.8)
ax.grid(which='minor', color='#EEEEEE', linestyle='-', linewidth=0.5)
ax.minorticks_on()
ax.set(xlabel='Deformacin Unitaria',ylabel='Esfuerzo fc [MPa]')
ax.set_title('Modelo Constitutivo Concreto: Popovics')
plt.legend()
```
